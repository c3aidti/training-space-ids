```python
import numpy as np
```

```python
interval = 0.1  # interval of loop in GB
N = int(128*1000*1000 * interval)   # number of float64s that fits in interval
r = np.random.rand(N)
```

```python
type(r[0])
```

```python
breaker = np.random.rand(1)
```

```python
flag = True
counter = 0
while(flag):
    breaker = np.concatenate((breaker,r), axis=0)
    counter = counter + 1
    print('Allocated', N*counter, 'float64s or', counter*interval, 'GB in memory')
        
```
