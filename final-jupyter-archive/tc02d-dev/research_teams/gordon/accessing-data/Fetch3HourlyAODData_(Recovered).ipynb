# 1. Collect a GeoSurfaceTimePoint

```python
"""
lat = -0.625
lon = -0.9375
time = "2017-07-15T06:00:00.000"

# filter
gstpFilter = c3.Filter().eq("latitude", lat).and_().eq("longitude", lon).and_().eq("time", time)

# lt or gt instead of eq for intervals

# collect instances 
gstp = c3.GeoSurfaceTimePoint.fetch({"filter": gstpFilter})
"""

lat1 = 0.0
lat2 = 10.0
lon1 = 0.0
lon2 = 10.0
time1 = "2017-07-01T00:00:00"
time2 = "2017-07-01T23:59:59"
gstpFilter = c3.Filter() \
    .ge("latitude", lat1) \
    .and_().le("latitude", lat2) \
    .and_().ge("longitude", lon1) \
    .and_().le("longitude", lon2) \
    .and_().ge("time", time1) \
    .and_().le("time", time2)
gstp = c3.GeoSurfaceTimePoint.fetch({"filter": gstpFilter})
```

```python
gstp
```

Good, we found it! Now we get the object itself to use it as a key.

```python
gstp = gstp.objs[0]
```

# 2. Use collected object as key to Simulation3HourlyAODOutput

```python
import pandas as pd
```

```python
# filter
s3haodFilter = c3.Filter().eq("geoSurfaceTimePoint", gstp) 

# collect objects
data = c3.Simulation3HourlyAODOutput.fetch({"filter": s3haodFilter})

# cast it into pandas
data = data.objs.toJson()
data = pd.DataFrame(data)
```

# 3. A weird observation?

We check how the fetched data over a time interval differ based on how the intervals are specified. Below we get around 
221 x 8 = 1768 rows of data for each day (which is what we expect) but only 2000 from both.

```python
lat = -0.625
lon = -0.9375
start_time = "2017-07-15T00:00:00.000"
end_time = "2017-07-15T21:00:00.000"

# filter
gstpFilter = c3.Filter().eq("latitude", lat).and_().eq("longitude", lon).and_().ge("time", start_time).and_().le("time", end_time)

# lt or gt instead of eq for intervals

# collect instances 
gstp = c3.GeoSurfaceTimePoint.fetch({"filter": gstpFilter}).objs

# filter
s3haodFilter = c3.Filter().eq("geoSurfaceTimePoint", gstp) 

# collect objects
data = c3.Simulation3HourlyAODOutput.fetch({"filter": s3haodFilter})

# cast it into pandas
data = data.objs.toJson()
data = pd.DataFrame(data)

print(data.shape)
```
