# Train Sudan, July, with dust

```python
lat1 = 8.0
lat2 = 8.5
lon1 = 38.0
lon2 = 38.5
time1 = "2017-07-01T00:00:00"
time2 = "2017-07-31T23:59:59"
gstpFilter = c3.Filter().ge("latitude", lat1).and_().le("latitude", lat2).and_().ge("longitude", lon1).and_().le("longitude", lon2).and_().ge("time", time1).and_().le("time", time2)
```

```python
c3.GeoSurfaceTimePoint.fetchCount({"filter": gstpFilter})
```

```python
GPR_kernel = c3.SklearnGPRKernelMatern(lengthScale=[1.0]*59, nu=0.5, coefficient=1.0).build().kernel

GPR_technique = c3.GaussianProcessRegressionTechnique(
                    randomState=42,
                    kernel = GPR_kernel
                )

jobOptions = c3.AODGaussianMLTrainingJobOptions(
    batchSize=1,
    gstpFilter=gstpFilter,
    targetName="dust",
    gprTechnique=GPR_technique
)

job = c3.AODGaussianMLTrainingJob(
    options=jobOptions
).upsert()
```

```python
job.start()
```

```python
job.status()
```

```python
job.id
```
