# Run individual models

### getFeatures

```python
featuresType = c3.TypeRef(typeName="SimulationModelParameters").toType()
fetchSpec = c3.FetchSpec(limit=-1)
inputTableC3= featuresType.fetch(fetchSpec).objs.toJson()
inputTablePandas = pd.DataFrame(inputTableC3)
inputTablePandas
```

### getTarget

```python
targetType = c3.TypeRef(typeName="Simulation3HourlyAODOutput").toType()
targetFilter = c3.Filter().eq("geoSurfaceTimePoint.id", "-0.625_-0.938_2017-07-01T00:20:00")
targetSpec = c3.FetchSpec(
    limit=-1,
    filter=targetFilter.toString()
)
outputTableC3 = targetType.fetch(targetSpec).objs.toJson()
outputTablePandas = pd.DataFrame(outputTableC3)
outputTablePandas
```

### Build new pipe and test

```python
GPR_kernel = c3.SklearnGPRKernelMatern(lengthScale=[1.0]*59, nu=0.5, coefficient=1.0).build().kernel

# define technique
GPReg_technique = c3.GaussianProcessRegressionTechnique(
                    randomState=42,
                    kernel = GPR_kernel
                )

# define features
featuresType = c3.TypeRef(typeName="SimulationModelParameters")
featuresSpec = c3.FetchSpec(limit=-1)

# define target
targetType = c3.TypeRef(typeName="Simulation3HourlyAODOutput")
targetFilter = c3.Filter().eq("geoSurfaceTimePoint.id", "-0.625_-0.938_2017-07-01T00:20:00")
targetSpec = c3.FetchSpec(
    limit=-1,
    filter=targetFilter.toString(),
    include="dust"
)

# create pipe
GPReg_pipe = c3.GaussianProcessRegressionPipe(
    technique=GPReg_technique,
    featuresType=featuresType,
    featuresSpec=featuresSpec,
    targetType=targetType,
    targetSpec=targetSpec
)
```

```python
X = GPReg_pipe.getFeatures()
dfX = c3.Dataset.toPandas(dataset=X)
dfX
```

```python
y = GPReg_pipe.getTarget()
dfy = c3.Dataset.toPandas(dataset=y)
dfy
```

### Train, save

```python
trainedGPR = GPReg_pipe.train(input=X, targetOutput=y)
```

```python
trainedGPR.upsert()
```

```python
c3.PythonSerialization.deserialize(serialized=trainedGPR.trainedModel.model).kernel_
```

# Run as batch job

- Construct the filter that defines the `GeoSurfaceTimePoint`s over which the models will be trained.
- Fetch it to see how many models will be trained.

```python
lat = -0.625
time = "2017-07-01T12:20:00"
gstpFilter = c3.Filter().eq("latitude", lat).and_().eq("time", time)
```

```python
c3.GeoSurfaceTimePoint.fetchCount({"filter": gstpFilter})
```

- Build the model technique. Each model will be trained with the same technique,

```python
GPR_kernel = c3.SklearnGPRKernelMatern(lengthScale=[1.0]*59, nu=0.5, coefficient=1.0).build().kernel

GPR_technique = c3.GaussianProcessRegressionTechnique(
                    randomState=42,
                    kernel = GPR_kernel
                )
```

- Define the job options.

```python
jobOptions = c3.AODGaussianMLTrainingJobOptions(
    batchSize=1,
    gstpFilter=gstpFilter,
    targetName="dust",
    gprTechnique=GPR_technique
)
```

- Create the job object and save it to table.
- Start the job.

```python
job = c3.AODGaussianMLTrainingJob(
    options=jobOptions
).upsert()
```

```python
job.start()
```
