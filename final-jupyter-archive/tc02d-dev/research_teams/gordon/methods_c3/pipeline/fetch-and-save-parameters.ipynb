# Use an Emulator to define which models' parameters to collect

```python
myFeaturesNames = [
    "bparam",
    "acure_autoconv_exp_nd",
    "a_ent_1_rp",
    "acure_dry_dep_acc",
    "acure_dry_dep_ait",
    "acure_dry_dep_so2",
    "acure_bvoc_soa",
    "acure_dms",
    "acure_bl_nuc",
    "acure_ait_width",
    "acure_cloud_ph",
    "dbsdtbs_turb_0",
    "acure_bc_ri",
    "acure_sea_spray",
    "acure_carb_bb_diam",
    "acure_anth_so2_r",
    "acure_prim_so4_diam"
]

a_init = Emulator()
```

```python
a = Emulator(
    region=[-45, 40, -30, 10],
    excludeFeats=list(set(a_init.__get_input_names__()) - set(myFeaturesNames)),
    start_time="2017-07-04T09:00:00",
    end_time="2017-07-04T13:59:59",
    centerTarget=True
)
```

# Start the MapReduce job to collect

```python
job = a.__retrieve_emulator_pixelwise__()
```

```python
job.status()
```

# Define a Clusterer to tabulate the parameters and save them to a file

```python
b = Clusterer(a, job)
```

```python
b.lengthscales
```

```python
with open("length_scale_tables/2017-07-03T09:00:00_2017-07-03T13:59:59", "w") as f:
    f.write(b.lengthscales.to_csv())
```
