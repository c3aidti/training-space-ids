# Direct Insertion

```python
feats = []

for i in range(10):
    x = {}
    for j in range(5):
        name = "var" + str(j)
        x[name] = np.random.rand()
        
    feat = c3.StagedFeatures(
        id=str(i+1),
        features=x
    )
    feats.append(feat)
    
c3.StagedFeatures.upsertBatch(feats)
```

```python
targets = []

for i in range(10):
    x = {}
    x["var0"] = np.random.rand()
 
    trgt = c3.StagedTargets(
        id=str(i+1),
        targets=x
    )
    targets.append(trgt)
    
c3.StagedTargets.upsertBatch(targets)
```
